```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt

# Generate synthetic data
x = np.linspace(0, 1, 100)
y = 2 * x + np.random.normal(0, 0.1, x.shape) + 1

# Define the neural network model
model = Sequential()
model.add(Dense(1000, input_dim=1, activation="linear"))
model.add(Dense(1000, activation="linear"))
model.add(Dense(1))

# Compile the model
model.compile(
    optimizer=tf.keras.optimizers.SGD(learning_rate=0.1), loss="mean_squared_error"
)

# Train the model
model.fit(x, y, epochs=100, batch_size=10)
# Predict using the model
predictions = model.predict(x)

model.summary()

plt.figure(figsize=(10, 6))
plt.scatter(x, y, label="Original Data")
plt.plot(x, predictions, color="red", label="Predictions")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Original Data vs Predictions")
plt.legend()
plt.show()
```

```python
# Generate new synthetic data using sine function
y_sine = np.sin(2 * np.pi * x)

# Define the neural network model
model_sine = Sequential()
model_sine.add(Dense(1000, input_dim=1, activation="relu"))
model_sine.add(Dense(1000, activation="relu"))
model_sine.add(Dense(1))

# Compile the model
model_sine.compile(
    optimizer=tf.keras.optimizers.SGD(learning_rate=0.1), loss="mean_squared_error"
)

# Train the model
model_sine.fit(x, y_sine, epochs=100, batch_size=10)

# Predict using the model
predictions_sine = model_sine.predict(x)

plt.figure(figsize=(10, 6))
plt.scatter(x, y_sine, label="Original Data (Sine)")
plt.plot(x, predictions_sine, color="red", label="Predictions")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Original Data (Sine) vs Predictions")
plt.legend()
plt.show()
```
